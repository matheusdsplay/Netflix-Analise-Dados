```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/netflix_titles.csv")

paises = df['country'].dropna().str.split(',').explode().str.strip()

traducao = {
    "United States": "Estados Unidos",
    "United Kingdom": "Reino Unido",
    "Canada": "Canadá",
    "India": "Índia",
    "Japan": "Japão",
    "South Korea": "Coreia do Sul",
    "France": "França",
    "Spain": "Espanha",
    "Germany": "Alemanha",
    "Brazil": "Brasil"   
}

paises_traduzidos = paises.map(lambda x: traducao.get(x, x))

top_paises = paises_traduzidos.value_counts().head(10)

top_paises.plot(kind='bar', figsize=(10,5), color='orange')
plt.title("Top 10 países com mais títulos")
plt.xlabel("País")
plt.ylabel("Quantidade de títulos")
plt.show()
```
