```python
# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o CSV
df = pd.read_csv("../data/netflix_titles_limpo.csv") 

# Filtrar lançamentos entre 2012 e 2025
df_filtrado = df[(df['release_year'] >= 2012) & (df['release_year'] <= 2025)]

# Contar lançamentos por ano e plotar
df_filtrado['release_year'].value_counts().sort_index().plot(kind="line", figsize=(10,5))
plt.title("Lançamentos entre 2012 e 2025")
plt.xlabel("Ano")
plt.ylabel("Quantidade de títulos")
plt.grid(True)
plt.show()
```
